```python
import requests
from bs4 import BeautifulSoup
import csv
import time
```

```python
r=requests.get("https://www.booking.com/hotel/es/libere-madrid-chamberi.es.html#tab-reviews")
html=BeautifulSoup(r.text,"lxml")
```

```python
html = BeautifulSoup(r.text, "lxml")

reviews = html.select(".review_item_content")

for rev in reviews:
    texto = rev.get_text(strip=True)
    print("----- REVIEW -----")
    print(texto)

print(r.status_code)
```

```python
html = BeautifulSoup(html, "html.parser")

body = html.body                      # obtener el <body>
divs_en_body = body.find_all("div")   # todos los <div> dentro del body

# div número 22 → índice 21
if len(divs_en_body) >= 22:
    div_22 = divs_en_body[21]
    print(div_22)
else:
    print("No hay 22 divs dentro de <body>")
```
